```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/spotify-2023.csv', delimiter = ',')

df
```

```python
cols_to_drop = ['in_apple_playlists', 'in_apple_charts', 'in_deezer_charts','in_shazam_charts']

df.drop(columns=cols_to_drop, inplace=True)

df.columns
```

```python
df.columns
```

```python

# Filtrar las filas donde 'released_year' es 2023
df = df[df['released_year'] == 2023]

# Mostrar la data
df
```

```python
import altair as alt


# Generar gráfico de barras
mark_bar = alt.Chart(df).mark_bar().encode(
    x=alt.X('streams:Q', title='Reproducciones'),
    y=alt.Y('track_name:N', sort='-x', title='Nombre de la Canción'),
    color=alt.Color('artist(s)_name:N', legend=alt.Legend(title="Artista"))
).properties(
    title="Canciones más reproducidas en Spotify en 2023"
)

# Imprimir gráfico de barras
mark_bar
```

```python
mark_bar.save('/content/drive/MyDrive/Colab Notebooks/mark_bar.html')
```

```python
# Generar gráfico de puntos
mark_circle = alt.Chart(df).mark_circle().encode(
    x=alt.X('bpm:Q', title='Beats Por Minuto (BPM)'),
    y=alt.Y('energy_%:Q', title='Energía (%)'),
    color=alt.Color('artist(s)_name:N', legend=alt.Legend(title="Artista")),
    tooltip=['track_name', 'artist(s)_name', 'bpm', 'energy_%']
).properties(
    title="Relación entre BPM y Energía de las Canciones en 2023"
)

# Imprimir gráfico de puntos
mark_circle
```

```python

# Agrupar el DataFrame por mes y día, y contar el número de canciones
grouped_df = df.groupby(['released_month', 'released_day']).size().reset_index(name='count')

# Generar gráfico de burbuja
bubble_chart = alt.Chart(grouped_df).mark_circle().encode(
    x=alt.X('released_day:O', title='Día del Mes'),
    y=alt.Y('released_month:O', title='Mes de Lanzamiento'),
    size=alt.Size('count:Q', title='Número de Canciones'),
    color=alt.Color('count:Q', scale=alt.Scale(scheme='greys'), legend=alt.Legend(title="N° Canciones")),
    tooltip=['released_month', 'released_day', 'count']
).properties(
    title="Número de Canciones lanzadas por Día y Mes en 2023",
    width=900
)

# Imprimir el gráfico de burbuja
bubble_chart
```

```python
bubble_chart.save('/content/drive/MyDrive/Colab Notebooks/bubble_chart.html')
```

```python
# Generar gráficos de barras de conteo de canciones por Clave y Modo
mark_bar_key_mode = alt.Chart(df).mark_bar().encode(
    x=alt.X('key:O', title='Clave'),
    y=alt.Y('count():Q', title='Número de Canciones'),
    color='mode:N',
    tooltip=['key', 'mode', 'count()']
).properties(
    title="Número de Canciones por Clave y Modo"
).facet(
    column='mode:N'
)

# Imprimir los gráficos de barra
mark_bar_key_mode
```

```python
# Crear un nuevo DataFrame basado en df, dividir la columna de nombres de artistas si hay más de uno
df_artists = df.copy()
df_artists['artist(s)_name'] = df_artists['artist(s)_name'].str.split(', ')
df_exploded = df_artists.explode('artist(s)_name')

# Convertir 'streams' a int y sumar por artista
df_exploded['streams'] = df_exploded['streams'].astype(int)
df_artist_streams = df_exploded.groupby('artist(s)_name').streams.sum().reset_index()

# Gráfico
mark_bar_artistas = alt.Chart(df_artist_streams).mark_bar().encode(
    x='streams:Q',
    y=alt.Y('artist(s)_name:N', sort='y'),
    tooltip=['artist(s)_name', 'streams']
).properties(title="Streams by Artist")

mark_bar_artistas
```
